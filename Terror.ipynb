```python
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
filename = '/content/Terror_Incident_Dataset.csv'
names = ['ID','Date','Day_of_Week','Time','Location','Attack_Type','Perpetrator','Known_Associates','Victims_Injured']
dataframe = read_csv(filename, names=names)
array = dataframe.values
X = array[:,0:8]
Y = array[:,8]
print(dataframe)
seed = 7
kfold = KFold(n_splits=10, random_state=seed, shuffle=True)
cart = DecisionTreeClassifier()
num_trees = 100
model = BaggingClassifier(estimator=cart, n_estimators=num_trees, random_state=seed)
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# Load Dataset
df = pd.read_csv("/content/Terror_Incident_Dataset.csv")

# Selecting Features and Target Variable
features = ['Victims_Injured', 'Victims_Deceased', 'Known_Associates', 'Operatives_Captured', 'Major_Incident']
X = df[features]
Y = df['Operational_Success'].astype('category').cat.codes  # Encoding categorical target variable

# Apply Recursive Feature Elimination (RFE)
model = LogisticRegression(max_iter=400)
rfe = RFE(estimator=model, n_features_to_select=3)
fit = rfe.fit(X, Y)

# Print Selected Features
selected_features = [features[i] for i in range(len(features)) if fit.support_[i]]
print("Num Features: %d" % fit.n_features_)
print("Selected Features:", selected_features)
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import MinMaxScaler
from numpy import set_printoptions

# Load the dataset
file_path = "/content/Terror_Incident_Dataset.csv"
df = pd.read_csv(file_path)

# Selecting only numerical features for chi-square test
numeric_features = ['Known_Associates', 'Victims_Injured', 'Victims_Deceased', 'Operatives_Captured', 'Major_Incident']
X = df[numeric_features]
Y = df['Major_Incident']  # Target variable

# Replace negative or invalid values with absolute values (if applicable)
X = X.abs()

# Apply Min-Max Scaling
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Apply SelectKBest with chi-square test
k = 4  # Select top 4 features
test = SelectKBest(score_func=chi2, k=k)
fit = test.fit(X_scaled, Y)

# Print feature scores
set_printoptions(precision=3)
print("Feature Scores:", fit.scores_)

# Select the top k features
selected_features = fit.transform(X_scaled)
print("Selected Features Shape:", selected_features.shape)
```
