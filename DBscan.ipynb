```python
import pandas as pd
file_path = "/content/Wholesale customers data.csv"
df = pd.read_csv(file_path)
```

```python
print(df.info())
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=6)
dbscan.fit(df)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([df,cl],axis=1)
```
